```python
# 学習モデルによる単独漢字識別
# POSITIONのデータ＋識別漢字上位3つ

#1 ライブラリのインポート等

from keras.models import load_model
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import glob

kanji_filename = "sakuma-0035-0020"
kanji_PATCHES = "image-PATCHES/content/image-PATCHES/"+kanji_filename+"/"
kanji_POSITION = "kanjiPATH/content/kanjiPATH/"+kanji_filename+"_POSITIONS.txt"
title_dict = {}

files = glob.glob(kanji_PATCHES + "/*png")
files.sort()

#2 各種設定

image_width = 64   # ここを変更
                  # 利用する学習済みモデルの横の幅のピクセル数と同じにする
image_height = 64  # ここを変更
                  # 利用する学習済みモデルの縦の高さのピクセル数と同じにする
color_setting = 1  # ここを変更。利用する学習済みモデルのカラー形式と同じにする
                  # 「1」はモノクロ・グレースケール。「3」はカラー。


#3 各種読み込み

model = load_model('model/model-64-20-500.h5')  # ここを変更

train_data_path = '147kanji/*' # ここを変更。
folder = glob.glob(train_data_path)
folder =  sorted(folder)

kanji_POSITION_list = []
with open(kanji_POSITION, 'r') as file:
    lines = file.readlines()
    for line in lines:
      kanji_POSITION_list.append(line)


POSITION_save = 'position/' + kanji_filename + '_POSITION-kanji.txt'  # 保存用ファイル

# 一時的な結果を保存するリストを作成
results = []
num=0
#4 画像の表示・各種設定等
for file in files:
    if color_setting == 1:
      img = cv2.imread(file, 0)
    elif color_setting == 3:
      img = cv2.imread(file, 1)
    img = cv2.resize(img, (image_width, image_height))
    plt.imshow(img)
    if color_setting == 1:
      plt.gray()
      plt.show()
    elif color_setting == 3:
      plt.show()

    print(file)
    img = img.reshape(image_width, image_height, color_setting).astype('float32')/255


    #5 予測と結果の表示等

    prediction = model.predict(np.array([img]))
    result = prediction[0]

    kanji_prediction = {}
    for i, accuracy in enumerate(result):
      if int(accuracy * 100) != 0:
        print('画像認識AIは「', os.path.basename(folder[i]), '」の確率を', int(accuracy * 100), '% と予測しました。')
        if int(accuracy * 100) < 10:
          kanji_prediction[os.path.basename(folder[i])] = "0"+str(int(accuracy * 100))+"%"
        else:
          kanji_prediction[os.path.basename(folder[i])] = str(int(accuracy * 100))+"%"

    kanji_prediction_02 = sorted(kanji_prediction.items(), key = lambda x : x[1], reverse=True)

    # 結果を一時的に保存する
    result_text = kanji_POSITION_list[num] + str(kanji_prediction_02[0]) + "\n"
    if len(kanji_prediction_02) > 1:
        result_text += str(kanji_prediction_02[1]) + "\n"
    else: result_text += "\n"
    if len(kanji_prediction_02) > 2:
        result_text += str(kanji_prediction_02[2]) + "\n"
    else: result_text += "\n"
    results.append(result_text)

    print('-------------------------------------------------------')
    num += 1

# ファイルにまとめて書き込む
with open(POSITION_save, 'w') as file:
    file.writelines(results)

print("finish")
```

```python
# 「今」「有」「圖」「如」の予測から正解率を上げる

file_path = "position/" + kanji_filename + "_POSITION-kanji.txt"
group_size = 4 # POSITION + kanji3つ
keywords = ["今", "有", "圖", "如"]

# 文章を取得
with open(file_path, "r") as file:
    lines = file.readlines()

# 摘出
result = []
for i in range(0, len(lines), group_size):
    group_of_lines = [line.strip("\n") for line in lines[i:i+group_size]]
    if any(keyword in line for keyword in keywords for line in group_of_lines):
        result.append(group_of_lines)

# output
output_file_path = "position/"+kanji_filename+"_output.txt"
with open(output_file_path, "w") as output_file:
    for group in result:
        for line in group:
            output_file.write(line + "\n")
        output_file.write("\n")

print("finish")
```

```python
# outputから予測できた座標から図形が含まれているか判別

import cv2

# 初期設定
file_path = "position/" + kanji_filename + "_POSITION-kanji.txt"
group_size = 4 # POSITION + kanji3つ
image = "image-median-3/content/image-median-3/" + kanji_filename + ".png"

# 文章を取得
with open(file_path, "r") as file:
    lines = file.readlines()

# 摘出
params_A = []
kanjis_A = []
base_x = 0
for i in range(0, len(lines), group_size):
    for j, line in enumerate(lines[i:i+group_size]):
        if j == 0:
            y = int(line.split(" ")[1]) # y座標
            x = int(line.split(" ")[2]) # x座標
            r = float(line.split(" ")[3]) # 半径
            param = y,x,r
        if "今" in line or "有" in line or "如" in line or "圖" in line:
            percent = line.split(",")[1]
            percent = percent.split("%")[0]
            percent = percent.split("'")[1]
            if int(percent) >= 80:
                if 900 <= int(x) <= 1200 and int(y) < 1100:
                    params_A.append(param)
                    kanjis_A.append(line.split("\n")[0])
                    if "有" in line and r > 28 and int(percent) > 98:
                        base_x = int(x)

# 「有」を基準に17px（≒40％）以内の位置なら採用
n=0
params_B = []
kanjis_B = []
if base_x != 0:
    for y,x,r in params_A:
        if base_x-17 <= int(x) <= base_x+17:
            params_B.append(params_A[n])
            kanjis_B.append(kanjis_A[n])
        n+=1

# 4文字含まれているか判別
kanjis_C = []
for KANJI in kanjis_B:
    kanjis_C.append(KANJI.split("'")[1])

image = cv2.imread("image-median-3/content/image-median-3/" + kanji_filename + ".png")
for y,x,r in params_B:
    cv2.rectangle(image, (int(x-r), int(y-r)), (int(x+r), int(y+r)), (0, 0, 255), 2)

image_name = "image_blob/" + kanji_filename + ".png"
cv2.imwrite(image_name,image)

print(kanjis_B)
print(params_B)
if len(kanjis_B) == 4:
    if "今" in kanjis_C and "有" in kanjis_C and "如" in kanjis_C and "圖" in kanjis_C:
        print(kanji_filename + ".pngに100%の確率で図形があります。")
elif len(kanjis_B) == 3:
    print(kanji_filename + ".pngに75%の確率で図形があるかもしれません。")
elif len(kanjis_B) == 2:
    print(kanji_filename + ".pngに50%の確率で図形があるかもしれません。")
elif len(kanjis_B) == 1:
    print(kanji_filename + ".pngに25%の確率で図形があるかもしれません。")
elif len(kanjis_B) > 5:
    print(kanji_filename + ".pngを確認してください。")
else:
    print(kanji_filename + ".pngに図形がありません。")
```
